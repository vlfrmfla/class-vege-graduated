```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# seaborn의 내장 데이터셋을 불러옵니다.
titanic_data = sns.load_dataset('titanic')
```

```python
# 생존자 수 시각화
sns.countplot(x='survived', data=titanic_data)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# 성별에 따른 생존율
sns.barplot(x='sex', y='survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#### 3.1. 생존율 계산

# 전체 생존율
survival_rate = titanic_data['survived'].mean()
print(f'Overall Survival Rate: {survival_rate:.2f}')
```

```python
# 나이 분포
sns.histplot(titanic_data['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 성별에 따른 생존율
gender_survival_rate = titanic_data.groupby('sex')['survived'].mean()
print(gender_survival_rate)

# 클래스에 따른 생존율
class_survival_rate = titanic_data.groupby('class')['survived'].mean()
print(class_survival_rate)
```
